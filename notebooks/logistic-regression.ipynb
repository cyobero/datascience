```python
df = pd.read_stata('./datascience/datasets/coke.dta')
df.head()
```

`pratio` is the ratio of coke price to pepsi price

```python
df.describe()
```

```python
X = np.array(df['pratio']).reshape(-1, 1)
y = np.array(df['coke']).reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train
```

```python
model = LogisticRegression()
fit = model.fit(X_train, y_train)
```

Evaluate our model...

```python
y_hat = fit.predict(X_test)
loss = log_loss(y_test, y_hat)
acc = accuracy_score(y_test, y_hat)

print("log-loss: {}".format(loss))
print("accuracy: {}".format(acc))
```

```python
plt.plot
```

```python
y_hat_prob = fit.predict_proba(X_test)
y_hat_prob = y_hat_prob[:, 0]
```

```python
plt.scatter(X_test, y_hat_prob)
plt.xlabel("Pepsi/Coke Price Ratio")
plt.ylabel("Predicted Probability")
plt.title("Estimated Probability of Choosing Coke Over Pepsi")
plt.show()
```
